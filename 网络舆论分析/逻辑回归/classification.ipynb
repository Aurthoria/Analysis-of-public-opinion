```python
#卡方检验
def chi_square_test(word_id,topic_id):
    A,B,C,D =0,0,0,0
    i=0
    for Title in json3:
        if str(word_id) in json3[Title] and y[i][topic_id] == 1:
            A+=1
        if str(word_id) in json3[Title] and not(y[i][topic_id] == 1):
            B+=1
        if not(str(word_id) in json3[Title]) and y[i][topic_id] == 1:
            C+=1
        if not(str(word_id) in json3[Title]) and not(y[i][topic_id] == 1):
            D+=1
        i+=1
    N = 1000
    res = N*(A*D-B*C)**2/((A+C)*(B+D)*(A+B)*(C+D))
    return res
```

```python
#将数据集分为训练集、测试集、验证集、小训练集
training_set=[X[0:800],Y[0:800]]
testing_set=[X[800:1000],Y[800:1000]]
validation_set=[[X[0:160],Y[0:160]],
                [X[160:320],Y[160:320]],
                [X[320:480],Y[320:480]],
                [X[480:640],Y[480:640]],
                [X[640:800],Y[640:800]]]
straining_set=[[X[160:800],Y[160:800]],
               [np.concatenate((X[0:160],X[320:800]),axis=0),np.concatenate((Y[0:160],Y[320:800]),axis=0)],
               [np.concatenate((X[0:320],X[480:800]),axis=0),np.concatenate((Y[0:320],Y[480:800]),axis=0)],
               [np.concatenate((X[0:480],X[640:800]),axis=0),np.concatenate((Y[0:480],Y[640:800]),axis=0)],
               [X[0:640],Y[0:640]]]
```

```python
#分类类，用静态方法
class Classify(object):
    @staticmethod
    def softmax(x):
        if len(x.shape)==2:
            max_list=[]
            for i in range(x.shape[1]):
                max_list.append(np.max(x[:,i]))
            x = x - np.array(max_list)
            exp_x = np.exp(x)
            softmax_x = exp_x / sum(exp_x)
        else:
            x = x - np.max(x)
            exp_x = np.exp(x)
            softmax_x = exp_x / sum(exp_x)
        return softmax_x
    @staticmethod
    def train(training_set,testing_set,lamda,alpha=1):
        max_epoch=1000
        W=np.random.randn(4,101)
        num=training_set[0].shape[0]
        for epoch in  range(max_epoch):
            grad=np.zeros([4,101])
            for k in range(num):
                #k=np.random.randint(0,num)
                output=Classify.softmax(np.dot(W,training_set[0][k].reshape(101,1)))
                grad=(output-np.array([training_set[1][k]]).T)*training_set[0][k]
                grad+=2*lamda*W
                W-=alpha*grad
        return W,Classify.evaluation(testing_set,W)
    def evaluation(testing_set,W):
        #评估函数
        num=testing_set[0].shape[0]
        confusion_matrix=[[0 for i in range(4)] for  j in range(4)]
        for i in range(num):
            predict=np.argmax(Classify.softmax(np.dot(W,testing_set[0][i])))
            confusion_matrix[np.argwhere(testing_set[1][i] == 1)[0][0]][predict]+=1
        confusion_matrix = np.array(confusion_matrix)
        TP = np.array([confusion_matrix[i,i] for i in range(4)])
        precision,recall =[],[]
        for i in  range(4):
            if sum(confusion_matrix.T)[i]==0:
                precision.append(1)
            else:
                precision.append(TP[i]/sum(confusion_matrix.T)[i])
        for i in  range(4):
            if sum(confusion_matrix)[i]==0:
                recall.append(1)
            else:
                recall.append(TP[i]/sum(confusion_matrix)[i])
        precision=np.array(precision)
        recall=np.array(recall)
        F_score=2*precision*recall/(precision+recall)  
        F_score,precision,recall =list(F_score),list(precision),list(recall)
        return  F_score,precision,recall  
```

```python
#交叉验证
F_scores=[]
for lamda in [0.1*i for i in range(1,5)]:
    F_score=[]
    for j in range(5):
        W1,K=Classify.train(straining_set[j],validation_set[j],lamda)
        F_score.append(np.mean(K[1]))
    F_scores.append(F_score)
#对五个训练集、验证集的误差取平均值
F_scores=list(map(np.mean,F_scores))
Lambda=F_scores.index(max(F_scores))*0.1+0.1#取F_score最大对应的正则化系数   
```
